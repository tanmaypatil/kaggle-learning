```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = {'A' : [ 1 , 2 , 3, 4 , 1, 1, 5,6,4] }
```

```python
df = pd.DataFrame(data)
```

```python
df
```

```python
# histogram 
sns.histplot(data=df, x='A', bins='auto', kde=False)
```

```python
# Rug Plot
sns.rugplot(data=df, x='A')
```

```python
# box Plot
sns.boxplot(data=df, x='A')
```

```python
ECDF (Empirical Cumulative Distribution Function) Plot:
sns.ecdfplot(data=df, x='A')
```

```python
#violin Plot:
sns.violinplot(data=df, x='A')
```

```python
sns.countplot(data=df, x='A')
```

```python
sns.kdeplot(data=df, x='A')
```
